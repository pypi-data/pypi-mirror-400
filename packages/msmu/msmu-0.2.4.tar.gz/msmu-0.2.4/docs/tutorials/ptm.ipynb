# Post-Translational Modification (PTM)

```python
# Import required packages:
import msmu as mm
```

```python
mdata = mm.read_h5mu("path/to/mdata.h5mu")
global_mdata = mm.read_h5mu("path/to/global_mdata.h5mu")
```

```python
mdata = mm.pp.infer_protein(mdata, propagated_from=global_mdata)
```

```python
mdata = mm.pp.to_ptm(mdata, modi_name="phospho", modification="[+79.633]")
```

```python
mdata = mm.pp.adjust_ptm_by_protein(
    mdata,
    modi_name="phospho",
    global_mdata=global_mdata,
    method="ridge",
    rescale=True,
)
```
