```python
from tqdm import tqdm
from time import sleep
from yaspin import yaspin
import logging

logger = logging.getLogger(__name__)
```

```python
progressm = tqdm(
    total=5,
    unit="file",
    desc="main",
)

for m in range(5):
    sleep(0.1)
    progressm.update(1)

progressm.close()
```

```python
spinner = yaspin(text="Loading...")
spinner.start()
sleep(2)
spinner.stop()
```

```python
logger.info("done")
```
