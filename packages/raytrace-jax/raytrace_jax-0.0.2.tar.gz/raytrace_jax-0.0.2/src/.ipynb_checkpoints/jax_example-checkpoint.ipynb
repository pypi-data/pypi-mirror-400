```python
import jax
import sys
sys.path.append('..') #if running from repository
from raytrace_jax import sample_raytrace, sample_hamiltonian

def gaussian(params):
    return jax.scipy.stats.norm.logpdf(params).sum()

k,nk = jax.random.split(jax.random.PRNGKey(0), 2)
x0 = jax.random.normal(nk,shape=[10000])
chain,likelihood = sample_raytrace(key=k, params_init=x0, \
    log_prob_fn=gaussian, n_steps=1000, n_leapfrog_steps=10, \
    step_size=0.17, refresh_rate=0.0, metro_check=1, sample_hmc=False)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(chain[:,0], chain[:,1])
plt.title('Sampling Results')
plt.grid()
plt.show()
```

```python
import numpy

def target_gaussian(params):
    return jax.scipy.stats.norm.logpdf(params).sum()

def target_rosenbrock(x):
  return -(100.0*(x[1::2] - x[0::2]**2.0)**2.0 + (1 - x[0::2])**2.0).sum()

def target_cauchy(params):
    return jax.scipy.stats.t.logpdf(params, df=1).sum()

def save_chain(chain,filename, likelihood):
    skip = 1
    if ('cauchy' in filename): skip=1
    chain_np = numpy.array(chain[::skip])
    loss_np = numpy.array(likelihood[::skip])
    numpy.savetxt(filename,numpy.hstack((chain_np, loss_np[:, None])))


def run_pdf_tests(target_pdf, length_hmc, hmc_steps, length_rt, raytrace_steps, init_sd):
    #HMC
    name = target_pdf.__name__
    max_steps = 100
    print("HMC:")
    n_steps = 4000
    if (name=='target_rosenbrock'):
        n_steps = 10000
    for dims,step in hmc_steps.items():
        if (name=='target_cauchy'):
             n_steps = 200000
        k1,k2,nk = jax.random.split(jax.random.PRNGKey(0), 3)
        params_init = jax.random.normal(nk,shape=[int(dims)])*init_sd
        max_steps = int(length_hmc/step)
        chain,likelihood = sample_hamiltonian(k1, params_init, target_pdf, n_steps=4000, n_leapfrog_steps=max_steps, step_size=float(step))
        chain,likelihood = sample_hamiltonian(k2, chain[-1], target_pdf, n_steps=n_steps, n_leapfrog_steps=max_steps, step_size=float(step))
        save_chain(chain, "hmc_"+name+"_"+str(dims)+"d_"+str(max_steps)+"steps.dat",likelihood)
    print("\nRaytracing:")
    for dims,step in raytrace_steps.items():
        if (name=='target_cauchy'):
             n_steps = 200000
        k1,k2,nk = jax.random.split(jax.random.PRNGKey(0), 3)
        params_init = jax.random.normal(nk,shape=[int(dims)])*init_sd
        max_steps = int(length_rt/step)
        chain,likelihood = sample_raytrace(k1, params_init, target_pdf, n_steps=4000, n_leapfrog_steps=max_steps, step_size=float(step))
        chain,likelihood = sample_raytrace(k2, chain[-1], target_pdf, n_steps=n_steps, n_leapfrog_steps=max_steps, step_size=float(step))
        save_chain(chain, "raytrace_"+name+"_"+str(dims)+"d_"+str(max_steps)+"steps.dat",likelihood)
```

```python
run_pdf_tests(target_gaussian, 1.57, {2:0.67, 10:0.5, 100:0.36, 1000:0.21, 10000:0.15}, 
              1.57, {2:0.71, 10:0.71, 100:0.52, 1000:0.26, 10000:0.17}, 1)
```

```python
k1,k2,nk = jax.random.split(jax.random.PRNGKey(0), 3)
params_init = jax.random.normal(nk,shape=[2])
chain,likelihood = sample_hamiltonian(k1, params_init, target_cauchy, n_steps=4000, n_leapfrog_steps=91, step_size=0.5)
chain,likelihood = sample_hamiltonian(k2, chain[-1], target_cauchy, n_steps=10000, n_leapfrog_steps=91, step_size=0.5)

k1,k2,nk = jax.random.split(jax.random.PRNGKey(0), 3)
params_init = jax.random.normal(nk,shape=[2])
chain2,likelihood = sample_raytrace(k1, params_init, target_cauchy, n_steps=4000, n_leapfrog_steps=91, step_size=0.5)
chain2,likelihood = sample_raytrace(k2, chain2[-1], target_cauchy, n_steps=10000, n_leapfrog_steps=91, step_size=0.5)
```
