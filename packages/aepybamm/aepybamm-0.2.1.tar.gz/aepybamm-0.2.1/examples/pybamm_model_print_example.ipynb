```python
from aepybamm import get_params, pybamm_print_tools
```

```python
fp = "params/AE_gen1_BPX.json"

model_inputs = {
    "base": {},
    "SPMe": {"model_type": "SPMe"},
    "lumped thermal": {
        "htc_ext": 1,
        "extra_model_opts": {"thermal": "lumped"},
    },
}

for name, inputs in model_inputs.items():
    _, model = get_params(
        fp,
        **inputs,
    )

    print(f"Printing model equations: {name}\n---\n")
    print(pybamm_print_tools.as_string(model))
    print("\n")
```
