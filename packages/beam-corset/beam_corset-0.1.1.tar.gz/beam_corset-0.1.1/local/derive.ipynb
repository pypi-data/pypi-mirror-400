```python
import sympy as sp
```

```python
wl = sp.symbols('lambda', real=True, positive=True)
```

```python

w0 = sp.symbols('w0', real=True, positive=True)
focus_0, z0, z1 = sp.symbols('z_{f0} z_{0} z_{1}', real=True)

def zR(w0):
    return sp.pi * w0**2 / wl

A, B, C, D = sp.symbols('A B C D', real=True)

q0 = z0 - focus_0 + sp.I * zR(w0)
q0p = q0 + z1 - z0

vec_p0 = sp.Matrix([focus_0, w0])
q1 = ((A * q0p + B) / (C * q0p + D)).simplify()

focus_1 = z1 - sp.re(q1)
w1 = sp.sqrt(sp.im(q1) * wl / sp.pi)

params_0 = sp.Matrix([focus_0, w0])
params_1 = sp.Matrix([focus_1, w1])

jac = params_1.jacobian(params_0)
jac.simplify()
jac
```

```python
vec_q0p = sp.Matrix([sp.re(q0p), sp.im(q0p)])
jac_p0_q0 = vec_q0p.jacobian(vec_p0)
jac_p0_q0
```

```python
vec_q0p_sym = sp.Matrix([*sp.symbols('q0.real q0.imag', real=True)])
q0p_sym = vec_q0p_sym[0] + sp.I * vec_q0p_sym[1]

q1 = ((A * q0p_sym + B) / (C * q0p_sym + D)).simplify()
vec_q1 = sp.Matrix([sp.re(q1), sp.im(q1)])

jac_q0p_q1 = vec_q1.jacobian(vec_q0p_sym)
jac_q0p_q1.simplify()
# jac_q0p_q1[0,0]
# str(jac_q0p_q1[0,0].factor())

for i, j in [(0,0), (0,1), (1,0), (1,1)]:
    print(f"res[{i},{j}] = {str(jac_q0p_q1[i,j].factor())}")
```

```python
jac_q0p_q1[1,0].factor() == -jac_q0p_q1[0,1].factor()
```

```python
vec_q1_sym = sp.Matrix([*sp.symbols('q1.real q1.imag', real=True)])
q1_sym = vec_q1_sym[0] + sp.I * vec_q1_sym[1]

vec_p1 = sp.Matrix([z1 - sp.re(q1_sym), sp.sqrt(sp.im(q1_sym) * wl / sp.pi)])
jac_q1_p1 = vec_p1.jacobian(vec_q1_sym)
jac_q1_p1.simplify()
str(jac_q1_p1)
```

```python
z = sp.symbols('z', real=True)
waist = sp.symbols('self.waist', real=True, positive=True)
focus = sp.symbols('self.focus', real=True)
wavelength = sp.symbols('self.wavelength', real=True, positive=True)
rayleigh_range = sp.pi * waist**2 / wavelength
radius = waist * sp.sqrt(1 + ((z - focus) / rayleigh_range) ** 2)
res = radius.diff(sp.Matrix([focus, waist])).simplify()
display(res)
str(res)
```

```python
import sympy as sp
w0, f0, wl, z0, z1 = sp.symbols('waist focus wavelength z_{0} z_{1}', real=True)
q0 = z0 - f0 + sp.I * sp.pi * w0**2 / wl
q0p = q0 + z1 - z0 # propagate to just before element
print(sp.Matrix([sp.re(q0p), sp.im(q0p)]).jacobian(sp.Matrix([f0, w0])))
```

```python
import sympy as sp
q0r, q0i, A, B, C, D = sp.symbols('q0.real q0.imag A B C D', real=True)
q0 = q0r + sp.I * q0i
q1 = ((A * q0 + B) / (C * q0 + D)).simplify()
jac = sp.simplify(sp.Matrix([sp.re(q1), sp.im(q1)]).jacobian(sp.Matrix([q0r, q0i])))
assert jac[0,0].factor() == jac[1,1].factor()
assert jac[1,0].factor() == -jac[0,1].factor()
print("jac[0,0] =", jac[0,0].factor())
print("jac[0,1] =", jac[0,1].factor())
```

```python
import sympy as sp
q1r, q1i, z1, wl = sp.symbols('q1.real q1.imag z_{1} wavelength', real=True, positive=True)
f1 = z1 - q1r
w1 = sp.sqrt(q1i * wl / sp.pi)
print(sp.Matrix([f1, w1]).jacobian(sp.Matrix([q1r, q1i])))
```
